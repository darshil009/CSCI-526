```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('analytics.csv')
df = df[['Health 30', 'Health 60', 'Health 90', 'Health 120', 'Health 150', 'Health 180', 'Weights 30',
         'Weights 60', 'Weights 90', 'Weights 120', 'Weights 150', 'Weights 180', 'Completed']]
df = df.fillna(0)
df
```

```python
df1 = df[df['Completed'] == 1].mean()
```

```python
df2 = df[df['Completed'] == 0].mean()
```

```python
l = [['30', df1['Health 30'], df1['Weights 30']], ['60', df1['Health 60'], df1['Weights 60']], 
     ['90', df1['Health 90'], df1['Weights 90']], ['120', df1['Health 120'], df1['Weights 120']], 
     ['150', df1['Health 150'], df1['Weights 150']], ['180', df1['Health 180'], df1['Weights 180']]]
```

```python
df3 = pd.DataFrame(l,columns=['time', 'health', 'weights'])
print(df3)
df3 = df3.set_index(list(df3)[0])

df3
```

```python
_ = df3.plot( kind= 'bar' , secondary_y= 'weights' , rot= 0 )
plt.show()
```
